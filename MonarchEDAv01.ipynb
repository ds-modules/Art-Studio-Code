```python
#import libraries
import pandas as pd #for the famous dataframes df
from scipy.stats.stats import pearsonr #pearson correlation
import matplotlib.pyplot as plt #for graphics
%matplotlib inline
```

```python
source="monarch1997_2020.csv"
#all variables have data types
monarch=pd.read_csv(source, header=0)
monarch.fillna('null')
monarch.head(5)
```

```python
plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 6))

# Specify how our lines should look
ax.plot(monarch.Year, monarch.Count, color='tab:red', label='Observed')

# Same as above
ax.set_xlabel('Time')
ax.set_ylabel('Count')
ax.set_title('Monarch Abundance')
ax.grid(True)
ax.legend(loc='upper right');
plt.savefig("monarch.png", dpi=300)
plt.show()
```

```python
source="Hurricanes_Storms.csv"
#all variables have data types
storms=pd.read_csv(source, header=0)
storms.fillna('null')
storms.head(10)
```

```python
plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 6))

# Specify how our lines should look
ax.plot(storms.Year, storms.Storms, color='tab:blue', label='Storms')

# Same as above
ax.set_xlabel('Time')
ax.set_ylabel('Count')
ax.set_title('Storm Prevalence')
ax.grid(True)
ax.legend(loc='upper right');
plt.savefig("storms.png", dpi=300)
plt.show()
```

```python
from scipy.stats.stats import pearsonr #pearson correlation
r1 = pearsonr(monarch['Count'], storms['Storms'])
r2 = pearsonr(monarch['Count'], storms['Hurricanes'])
print(r1)
print(r2)
```

```python
myDualAxisgraph = storms['Storms'].plot(figsize=(10, 6), kind='line', title='Monarchs vs Storms', marker='o', fillstyle = 'full', color='#848f9b', linewidth=3);
myDualAxisgraph.set_ylabel('Storm Count', color='#848f9b');
myDualAxisgraph.spines['top'].set_visible(False)
myDualAxisgraph.tick_params('y', colors='#848f9b');

myDualAxisgraph2 = myDualAxisgraph.twinx();

myDualAxisgraph2 = monarch['Count'].plot(figsize=(10, 6), kind='line', marker='s', fillstyle = 'full', color='#841515', linewidth=1);
myDualAxisgraph2.spines['top'].set_visible(False)
myDualAxisgraph2.set_ylabel('Monarch Count', color='#841515');
myDualAxisgraph2.tick_params('y', colors='#841515');

fig = myDualAxisgraph2.get_figure()
fig.tight_layout()
fig.savefig('dual01.png', dpi=600)
```

```python
plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 6))

# Specify how our lines should look
ax.plot(storms.Year, storms.Storms, color='tab:blue', label='Storms')

# Same as above
ax.set_xlabel('Time')
ax.set_ylabel('Count')
ax.set_title('Storm Prevalence vs Monarch Abundance')
ax.grid(True)
ax.legend(loc='center right');

ax2=ax.twinx()
ax2.plot(monarch.Year, monarch.Count, color='tab:green', label='Monarchs')

# Same as above
ax2.set_xlabel('Time')
ax2.set_ylabel('Count')
ax2.legend(loc='center left');
plt.savefig("correlation.png", dpi=300)
plt.show()
```

```python
source="snow.csv"
#all variables have data types
snow=pd.read_csv(source, header=0)
snow.fillna('null')
snow.head()
```

```python
plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 6))

# Specify how our lines should look
ax.plot(snow.Year, snow.Sum, color='tab:blue', label='Snowpack')

# Same as above
ax.set_xlabel('Time')
ax.set_ylabel('mm')
ax.set_title('Snowpack Height')
ax.grid(True)
ax.legend(loc='upper right');
plt.savefig("snow.png", dpi=300)
plt.show()
```

```python
from scipy.stats.stats import pearsonr  

r1 = pearsonr(monarch['Count'], snow['Sum'])
print(r1)
```

```python
plt.rc('font', size=12)
fig, ax = plt.subplots(figsize=(10, 6))

# Specify how our lines should look
ax.plot(snow.Year, snow.Sum, color='tab:blue', label='Snow Pack')

# Same as above
ax.set_xlabel('Time')
ax.set_ylabel('Count')
ax.set_title('Snow Pack vs Monarch Abundance')
ax.grid(True)
ax.legend(loc='upper right');

ax2=ax.twinx()
ax2.plot(monarch.Year, monarch.Count, color='tab:green', label='Monarchs')

# Same as above
ax2.set_xlabel('Time')
ax2.set_ylabel('Count')
ax2.legend(loc='lower left');
plt.savefig("nocorrelation.png", dpi=300)
plt.show()
```
